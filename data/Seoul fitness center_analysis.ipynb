```python
import pandas as pd

data = pd.read_csv('서울시 체력단련장업 인허가 정보.csv', encoding='cp949')     # csv 파일 읽기
data.head()            # 상위 5개 확인
```

```python
data.tail()          # 하위 5개 확인
```

```python
data.columns
```

```python
df = pd.DataFrame(data)
df = df[[ '사업장명','상세영업상태명','지번주소','도로명주소']]
df
```

```python
# 영업중인 매장 sorting
df = df[df['상세영업상태명']=='영업중']
df
```

```python
# 인덱스 재정렬
df = df.reset_index(drop=True)
df
```

```python
# 지번 주소에서 구 parsing
address_split = df["지번주소"].str.split(" ")
df["지번(구)"] = address_split.str.get(1)
df
```

```python
# 도로명 주소에서 구 parsing
address_split = df["도로명주소"].str.split(" ")
df["도로명(구)"] = address_split.str.get(1)
df
```
