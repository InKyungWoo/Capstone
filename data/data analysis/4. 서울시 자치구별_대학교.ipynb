```python
# 서울시 대학교 데이터 가져오기
# 총 53개

import pandas as pd

data = pd.read_csv('../raw data/2022년도_공시대상학교목록.csv', encoding='UTF8')     # csv 파일 읽기
data.head()            # 상위 5개 확인
```

```python
# 하위 5개 확인
data.tail()
```

```python
# 자치구별 집계
grouped = data.groupby('자치구')['학교명'].count()
grouped
```

```python
sorted = grouped.sort_values(ascending=False)   # 많은 순서대로 sorting
sorted
```

```python
# 데이터 시각화

import matplotlib.pyplot as plt 

plt.rcParams['font.family'] = "AppleGothic"    # 한글 폰트 설정
plt.figure(figsize= (15,10))                    # 그래프의 사이즈
plt.bar(sorted.index, sorted)
plt.title('서울시 자치구별 대학교 개수')                 # 그래프의 제목
plt.xticks(rotation=0)
plt.show()                                     # 그래프 그리기
```
