```python
# 서울시 자치구별 직장인구 데이터 가져오기
# 총 34,483개의 레코드

import pandas as pd

data = pd.read_csv('../raw data/서울시 우리마을가게 상권분석서비스(상권-직장인구).csv', encoding='cp949')     # csv 파일 읽기
data.head()          # 상위 5개 확인
```

```python
data.tail()
```

```python
# 2022년 2분기 데이터 추출
data = data[data['기준_년월_코드']==2022]
data = data[data['기준_분기_코드']==2]
data

# 총 1,630개 레코드
```

```python
data.columns
```

```python
# 필요한 열 추출

data = data[['기준_년월_코드', '기준_분기_코드','상권_구분_코드_명','상권_코드_명','상권_구분_자치구별','총_직장_인구_수','연령대_20_직장_인구_수', '연령대_30_직장_인구_수']]
data
```

```python
# 자치구별 집계
# 20대 직장인구
grouped_20s = data.groupby('상권_구분_자치구별')['연령대_20_직장_인구_수'].sum()
grouped_20s
```

```python
# 자치구별 집계
# 30대 직장인구
grouped_30s = data.groupby('상권_구분_자치구별')['연령대_30_직장_인구_수'].sum()
grouped_30s
```

```python
# 타입 확인
type(grouped_20s)
```

```python
# 타입 변환 Series -> DataFrame
pd.DataFrame(grouped_20s)
```

```python
type(grouped_30s)
pd.DataFrame(grouped_30s)
```

```python
# 열 merge
grouped = pd.merge(grouped_20s, grouped_30s, on='상권_구분_자치구별')
grouped
```

```python
# csv 파일 저장
grouped.to_csv("working population.csv",encoding='utf8')
```

```python
import pandas as pd
data = pd.read_csv('working population.csv',encoding='utf8')
data
```

```python
# 합계 열 추가
sum = data.sum(axis=1)
```

```python
data['합계'] = sum
data
```

```python
# 인구수 합계 기준으로 정렬
sorted = data.sort_values(by=['합계'], ascending=False)
sorted
```

```python
# 데이터 시각화
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = "AppleGothic"             # 한글 폰트 설정

# 누적 막대 그래프로 나타내기 
ax = sorted.plot.bar(x='상권_구분_자치구별',y={'연령대_20_직장_인구_수','연령대_30_직장_인구_수'},stacked=True)    
ax.set_title('서울시 20,30대 직장인구 수')
plt.figure(figsize= (30,20))  
plt.show()
```
