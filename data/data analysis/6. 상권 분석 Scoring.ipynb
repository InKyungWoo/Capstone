```python
# 종합 파일 가져오기

import pandas as pd

data = pd.read_csv('종합_Scoring.csv', encoding='utf8')
data
```

```python
# 최대값 확인
max_working_pop = data['직장인구 합계'].max()
max_working_pop
```

```python
# 최대값 대비 자치구별 비율 구하기
data['직장인구_비율'] = data['직장인구 합계'] / max_working_pop
data
```

```python
# 대학교 비율 구하기

max_univ = data['대학교'].max()
data['대학교_비율'] = data['대학교'] / max_univ
data
```

```python
# 지하철역 비율 구하기

max_subway = data['지하철역'].max()
data['지하철역_비율'] = data['지하철역'] / max_subway
data
```

```python
# 헬스장 비율 구하기

max_gym = data['헬스장'].max()
data['헬스장_비율'] = data['헬스장'] / max_gym
data
```

```python
# 필요한 열 추출

data = data[['자치구별','직장인구_비율','대학교_비율','지하철역_비율','헬스장_비율']]
data
```

```python
# 기준 별 반영 비율 
# 직장인구 25%, 대학교 25%, 지하철역 10%, 헬스장 40%

data['직장인구 score'] = data['직장인구_비율'] * 25
data['대학교 score'] = data['대학교_비율']  * 25
data['지하철역 score'] = data['지하철역_비율'] * 10
data['헬스장 score'] = data['헬스장_비율'] * 40

data
```

```python
data = data[['자치구별','직장인구 score','대학교 score','지하철역 score','헬스장 score']]
data
```

```python
data['합계'] = data['직장인구 score'] + data['대학교 score'] + data['지하철역 score'] + data['헬스장 score']
data
```

```python
# csv 파일 저장
data.to_csv("Scoring 결과.csv",encoding='utf8')
```
