```python
import pandas as pd

data = pd.read_csv('서울시 체력단련장업 인허가 정보.csv', encoding='cp949')     # csv 파일 읽기
data.head()            # 상위 5개 확인
```

```python
data.tail()          # 하위 5개 확인
```

```python
# 전체 구분 목록 가져오기
data.columns
```

```python
# 필요한 열 추출
# 전체 매장 5477개
df = pd.DataFrame(data)
df = df[[ '사업장명','상세영업상태명','지번주소','도로명주소']]
df
```

```python
# 영업중인 매장 sorting, 총 3311개
df = df[df['상세영업상태명']=='영업중']
df
```

```python
# 인덱스 재정렬
df = df.reset_index(drop=True)
df
```

```python
# 지번 주소에서 구 parsing
address_split = df["지번주소"].str.split(" ")
df["지번(구)"] = address_split.str.get(1)
df
```

```python
# 도로명 주소에서 구 parsing
address_split = df["도로명주소"].str.split(" ")
df["도로명(구)"] = address_split.str.get(1)
df
```

```python
# 필요없는 열(주소) 제거
df = df[['사업장명','상세영업상태명','지번(구)','도로명(구)']]
df
```

```python
# 결손값 NAN 처리
import numpy as np
df.loc[df['지번(구)']=='','지번(구)'] = np.nan
df.loc[df['도로명(구)']=='','도로명(구)'] = np.nan
```

```python
# 지번 주소가 null인 경우   # 53개
df2 = df[df['지번(구)'].isnull()]
df2
```

```python
# 도로명(구) -> 지역(구)
df2.columns = [['사업장명','상세영업상태명','지번(구)','지역(구)']]
df2 = df2[['사업장명','상세영업상태명','지역(구)']]
df2
```

```python
# 도로명 주소가 null인 경우  # 63개
df3 = df[df['도로명(구)'].isnull()]
df3
```

```python
# 지번(구) -> 지역(구)
df3.columns = [['사업장명','상세영업상태명','지역(구)','도로명(구)']]
df3 = df3[['사업장명','상세영업상태명','지역(구)']]
df3
```

```python
# 결측값 있는 행 제거   # 3195개
df = df.dropna()
df
```

```python
# 필요 없는 열 제거
df.columns = [['사업장명','상세영업상태명','지역(구)','도로명(구)']]
df = df[['사업장명','상세영업상태명','지역(구)']]
df
```

```python
# df, df2(도로명 주소) 병합
df = pd.concat([df,df2], ignore_index=True)
df 
```

```python
# df, df3(지번 주소) 병합
df = pd.concat([df,df3], ignore_index=True)
df 
```

```python
# 열 제거
df = df[['사업장명','지역(구)']]
df
```

```python
# csv 파일 저장
df.to_csv("fitness.csv")
```

```python
# 지역별 집계
grouped = data.groupby('지역(구)')['사업장명'].count()
grouped
```

```python
# 데이터 시각화

import matplotlib.pyplot as plt 
sorted_grouped = grouped.sort_values(ascending=False)   # 많은 순서대로 sorting
sorted_grouped
```

```python
plt.rcParams['font.family'] = "AppleGothic"    # 한글 폰트 설정
plt.figure(figsize= (20,10))                    # 그래프의 사이즈
plt.bar(sorted_grouped.index, sorted_grouped)
plt.title('구 별 헬스장 현황')                 # 그래프의 제목
plt.xticks(rotation=0)
plt.show()                                     # 그래프 그리기
```
