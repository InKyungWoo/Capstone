```python
import csv
import pandas as pd

f = open('market.csv', encoding='UTF8')    # 저장된 파일 불러오기
data = csv.reader(f)
header = next(data)

seoul_gyeonggi, gangwon, chungcheong, jeolla, gyeonsang, jeju = 0, 0, 0, 0, 0, 0
cost_sg, cost_gw, cost_cc, cost_jl, cost_gs, cost_jj = 0,0,0,0,0,0

for row in data:
    region_do = row[1]    # 행정구역 도 별 분류
    row[4] = int(row[4])
    
    if region_do == '서울': 
        seoul_gyeonggi += 1
        cost_sg += row[4]
    elif region_do == '경기': 
        seoul_gyeonggi += 1
        cost_sg += row[4]
    elif region_do == '인천': 
        seoul_gyeonggi += 1
        cost_sg += row[4]
    elif region_do == '강원': 
        gangwon += 1
        cost_gw += row[4]
    elif region_do == '충남': 
        chungcheong += 1
        cost_cc += row[4]
    elif region_do == '충북': 
        chungcheong += 1
        cost_cc += row[4]
    elif region_do == '세종' : 
        chungcheong += 1
        cost_cc += row[4]
    elif region_do == '대전' : 
        chungcheong += 1
        cost_cc += row[4]
    elif region_do == '전남' : 
        jeolla += 1
        cost_jl += row[4]
    elif region_do == '전북' : 
        jeolla += 1
        cost_jl += row[4]
    elif region_do == '광주' : 
        jeolla += 1
        cost_jl += row[4]
    elif region_do == '경남' : 
        gyeonsang += 1
        cost_gs += row[4]
    elif region_do == '경북' : 
        gyeonsang += 1
        cost_gs += row[4]
    elif region_do == '부산' : 
        gyeonsang += 1
        cost_gs += row[4]
    elif region_do == '울산' : 
        gyeonsang += 1
        cost_gs += row[4]
    elif region_do == '대구' : 
        gyeonsang += 1
        cost_gs += row[4]
    elif region_do == '제주' : 
        jeju += 1
        cost_jj += row[4]

total = seoul_gyeonggi + gangwon + chungcheong + jeolla + gyeonsang + jeju 
avg_sg = cost_sg / seoul_gyeonggi
avg_gw = cost_gw / gangwon
avg_cc = cost_cc / chungcheong
avg_jl = cost_jl / jeolla
avg_gs = cost_gs / gyeonsang
avg_jj = cost_jj / jeju
avg = (cost_sg + cost_gw + cost_cc + cost_jl + cost_gs + cost_jj) / total

ary = {'지역' : ['서울/경기','강원','충청','전라','경상','제주','합계'],
      '점포 수' : [seoul_gyeonggi, gangwon, chungcheong, jeolla, gyeonsang, jeju, total],
      '평균 가격' : [f'{avg_sg:.2f}', f'{avg_gw:.0f}', f'{avg_cc:.2f}', f'{avg_jl:.0f}', f'{avg_gs:.0f}', f'{avg_jj:.0f}', f'{avg:.2f}']}
x = pd.DataFrame(ary, columns=['지역', '점포 수', '평균 가격'])
x
```

```python
import csv
import pandas as pd

f = open('market.csv', encoding='UTF8')    # 저장된 파일 불러오기
data = pd.read_csv('market.csv')

seoul_data = data[data['지역(시)']=='서울']   # 서울 지역 매장 sorting

groupdata = seoul_data.groupby('지역(구)')   # 구 별로 정리
market = groupdata['상호명']
seoul_count = market.count()                # 매장 수 count
seoul_count
```

```python
import matplotlib.pyplot as plt 

sorted_seoul_count = seoul_count.sort_values(ascending=False)   # 많은 순서대로 sorting

plt.rcParams['font.family'] = "AppleGothic"    # 한글 폰트 설정
plt.figure(figsize= (10,5))                    # 그래프의 사이즈
plt.bar(sorted_seoul_count.index, sorted_seoul_count)
plt.title('구 별 프로틴 카페 현황')                 # 그래프의 제목
plt.xticks(rotation=0)
plt.show()                                     # 그래프 그리기
```
